```python
import os
import pandas as pd

# 1. پہلے folder path الگ
folder_path = r'C:\Users\Shahrukh Tasleem\Downloads\t20_cricket_matches'

# 2. پھر file name الگ
file_name = 't20_matches.csv'  # یا جو بھی صحیح نام ہے

# 3. دونوں کو join کریں
file_path = os.path.join(folder_path, file_name)

if os.path.exists(file_path):
    data = pd.read_csv(file_path)  # ✅ اب صحیح ہے
    print("File found! ✅")
    print(data.head())
else:
    print("File not found! ❌")
    print("Downloads folder mein ye files hain:")
    print(os.listdir(folder_path))  # ✅ صرف folder path
    
print("\n=== basic data ===")
print(f"total rows: {len(data)}")
print(f"total columns : {len(data.columns)}")
print("\ncolumns:")
print(data.columns.tolist())

# 2. ڈیٹا کی ساخت دیکھیں
print("\n=== data information ===")
print(data.info())

# 3. پہلی 10 قطاریں دیکھیں
print("\n=== first 10 rows ===")
print(data.head(10))

# 4. خالی ویلیوز چیک کریں
print("\n=== null values ===")
print(data.isnull().sum())


# See what columns you have
print("\n=== All column names ===")
print(data.columns.tolist())

# Check the basic statistics
print("\n=== Basic statistics ===")
print(data.describe())



# Most common results
if 'result' in data.columns:
    result_counts = data['result'].value_counts()
    print("\n=== Match results summary ===")
    print(result_counts)


# Most winning teams
if 'team' in data.columns or 'winner' in data.columns:
    winning_teams = data['winner'].value_counts()
    print("\n=== Most successful teams ===")
    print(winning_teams.head(10))


# Date analysis (if you have date column)
if 'date' in data.columns:
    print("\n" + "="*40)
    print("Date analysis")
    print("="*40)
    
    #date column ko fix kro
    data['date'] = pd.to_datetime(data['date'], errors='coerce', dayfirst=True)

    #matches yearly
    data['year'] = data['date'].dt.year
    yearly_matches = data['year'].value_counts().sort_index()
    print("\n=== Matches per year ===")
    print(yearly_matches)
    print(f"\nEarliest matches: {data['date'].min()}")
    print(f"\nLatest matches: {data['date'].max()}")
    
print("\n" + "="*40)
print("match result analysis")
print("="*40)

if 'winner' in data.columns:
    print("top teams by matches won:")
    winners=data['winner'].value_counts().head(10)
    print(winners)

# Run analysis
if 'win_by_runs' in data.columns and 'win_by_wickets' in data.columns:
    win_by_runs=(data['win_by_runs']>0).sum()
    win_by_wickets=(data['win_by_wickets']>0).sum()
    other_wins= len(data) - win_by_runs - win_by_wickets
    
    print("\n=== Run statistics ===")
    print(f"Win by runs: {win_by_runs}")
    print(f"Win by wickets: {win_by_wickets}")
    print(f"other wins (tie\no result,etc): {other_wins}")    

print("\n"+ "="*40)
print("team performance")
print("="*40)


if 'home' in data.columns and 'away' in data.columns:
    home_wins=(data['winner']==data['home']).sum()
    away_wins=(data['winner']==data['away']).sum()
    print(f"home team wins: {home_wins}")
    print(f"away team wins: {away_wins}")
    print(f"home win percentage: {(home_wins/len(data))*100:.2f}%")
    
if 'player' in data.columns:
    top_players = data['player'].value_counts().head(10)
    print("\n=== Most frequent players ===")
    print(top_players)

print("\n"+"="*40)
print("innings analysis")
print("="*40)

if 'innings1_runs' in data.columns:
    print("first innings stats:")
    print(f"average 1st innings score: {data['innings1_runs'].mean():.2f}")
    print(f"highest 1st innings score: {data['innings1_runs'].max()}")
    print(f"lowest 1st innings score: {data['innings1_runs'].min()}")


if 'innings2_runs' in data.columns:
    print("second innings stats:")
    print(f"average 2st innings score: {data['innings2_runs'].mean():.2f}")
    print(f"highest 2st innings score: {data['innings2_runs'].max()}")
    print(f"lowest 2st innings score: {data['innings2_runs'].min()}")



print("\n"+"="*40)
print("venue analysis")
print("="*40)

if 'venue' in data.columns:
    print("to 10 venues by matches hosted: ")
    top_venues=data['venue'].value_counts().head(10)
    print(top_venues)


if 'D/L_method' in data.columns:
    print("D/L method usage")
    dl_matches= data['D/L_method'].value_counts()
    print(dl_matches)


print("\n" + "="*40)
print("overall summary:")
print("="*40)

print(f"total matches analyzed: {len(data)}")

if 'winner' in data.columns:
    print(f"matches with result: {data['winner'].notna().sum()}")
    print(f"without result: {data['winner'].isna().sum()}")


if 'win_by_runs' in data.columns:
    biggest_runs_win=data['win_by_runs'].max()
    print(f"biggest win by runs: {biggest_runs_win}")


if 'win_by_wickets' in data.columns:
    biggest_runs_wickets=data['win_by_wickets'].max()
    print(f"biggest win by wickets: {biggest_runs_wickets}")

    
# Save your cleaned data
data.to_csv('analyzed_cricket_data.csv', index=False)
print("\n=== Analysis saved successfully! ===")

print("\n=== My columns are ===", data.columns.tolist())

# 1. Match summary statistics
print("\n Total matches:", len(data))
print("Unique series:", data['series_id'].nunique())
print("\n Unique venues:", data['venue'].nunique())


# 2. Who wins most often?
winner_counts = data['winner'].value_counts()
print("\n=== Most successful teams ===")
print(winner_counts.head(10))

# 3. How do teams win?
win_by_runs = data['win_by_runs'].value_counts()
win_by_wickets = data['win_by_wickets'].value_counts()
print("\n=== Win methods summary ===")
print("Matches won by runs:", win_by_runs.sum())
print("Matches won by wickets:", win_by_wickets.sum())


 #4. Convert date to datetime and analyze trends
data['date'] = pd.to_datetime(data['date'],'dayfirst=True')
data['month'] = data['date'].dt.month_name()
data['year'] = data['date'].dt.year

# Matches per year
matches_per_year = data['year'].value_counts().sort_index()
print("\n=== Matches per year ===")
print(matches_per_year)



# 5. Performance by venue
venue_stats = data.groupby('venue').agg({
    'match_id': 'count',
    'innings1_runs': 'mean',
    'innings2_runs': 'mean'
}).rename(columns={'match_id': 'total_matches'})
print("\n=== Venue statistics ===")
print(venue_stats.sort_values('total_matches', ascending=False))


# 6. First vs second innings analysis
avg_first_innings = data['innings1_runs'].mean()
avg_second_innings = data['innings2_runs'].mean()
print(f"\nAverage 1st innings score: {avg_first_innings:.2f}")
print(f"Average 2nd innings score: {avg_second_innings:.2f}")

# Win percentage when batting first vs second
first_inn_wins = len(data[data['innings1'] == data['winner']])
second_inn_wins = len(data[data['innings2'] == data['winner']])
print(f"Matches won batting first: {first_inn_wins}")
print(f"Matches won batting second: {second_inn_wins}")

if 'scores' in data.columns:
    print("=== scores analysis ===")
    print("data type:", data['scores'].dtype)
    
    print("\nfirst 5 scores")
    for i, score in enumerate(data['scores'].head(),1):
        print(f"{i}.{score} (Type: {type(score)})")

    if data['scores'].dtype in ['int64', 'float64']:
        print(f"total scores: {data['scores'].sum()}")
        print(f" scores mean: {data['scores'].mean():.2f}")
        print(f" scores max: {data['scores'].max()}")
    else:
        print("can't calculate sum/mean/max- scores are having text format")
       # print("try u")
```
